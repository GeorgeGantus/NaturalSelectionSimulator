<h1>Simulação - Evolução de tentilhões</h1>

Com este projeto temos como objetivo realizar uma simulação da evolução dos tentilhões de maneira semelhante à identificada por darwin nas ilhas Galápago. Para isso foi necessário modelar alguns comportamentos e estruturas presentes na natureza a fim de simular a interação dos individios com o meio. Sendo assim começaremos explicando a estrutura do modelo.

<h3>Ambiente</h3>

O ambiente na qual os tentilhões se encontram foi modelado como sendo uma matriz 10x10 onde cada posição (x,y) é representada por um item da matriz. os individuos podem ocupar qualquer lugar da matriz. E a comida disponivel também está distribuida pelos pontos da matriz.

<h3>Comida</h3>

A fim de sobreviver e de se reproduzir os individuos devem buscar se alimentar durante o dia. Cada alimento consumido garante uma certa quantidade que varia de acordo com caracteristicas do individio e do proprio alimento. Uma quantidade fixa de comida é adicionada aleatoriamente a matriz a cada gereção. A comida é adicionada ao ambiente e é representada como um número inteiro em uma posição da matriz do ambiente, de modo que se a posição (2,3) possui o valor 2 quer dizer que existem duas comidas disponiveis naquele lugar. A comida possui um tamanho fixo, esse tamanho é utilzado para determinar quanto de energia um individuo obtem consumindo-a.

<h3>Individuo</h3>

O individuo representa um tentilhão na população, durante cada dia o individuo pode se mover a fim de procurar comida, quando ele se move para uma posição onde há alimento, ele consome 1 alimento daquela posição adquirindo uma quantidade de energia que varia de acordo com o tamanho da comida e o tamanho do bico daquele individuo. O tamanho do bico é uma caracteristica genética e que sofre mutação conforme os individuos se reproduzem. Quando o final do dia chega se a energia obtida por um individuo for inferior a certo limite ele morre. Se a energia for superior a um segundo limite ele se reproduz, gerando uma cópia de si mesmo com uma certa mutação aleatória. Caso a energia se encontre entre esses dois limites o individuo apenas sobrevive para mais um dia. 

<h3>População</h3>

A população é representada por um conjunto de individuos. Durante o dia eles se movem de maneira ordenada. A população varia ao longo das gerações com a morte e reprodução dos individuos.

<h2>Situação 1 - População adaptada</h2>

Nesta primeira situação representamos o caso onde os individuos já estão adaptados ao tamanho da comida de modo que o tamanho dos bicos inicialmente já é aquele que maximiza o ganho energético e o número de individuos é inicialmente proporciaonal à comida disponivel.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = open('../data.txt')
aux = file.read().strip().split("\n")
```

```python
data = [[float(y) for y in x.split(" ")] for x in aux]
```

```python
population = [x[0] for x in data[1:-1]]
food = [x[1] for x in data[1:-1]]
beak_size = [x[2] for x in data[1:-1]]
velocity = [x[3] for x in data[1:-1]]
```

```python
plt.plot([i for i in range(len(population))],population)
plt.xlabel('Generations')
plt.ylabel('População')
```

```python
plt.plot([i for i in range(len(food))],food)
plt.xlabel('Generations')
plt.ylabel('Comida Disponivel')
```

```python
plt.plot([i for i in range(len(beak_size))],beak_size)
plt.xlabel('Generations')
plt.ylabel('bico')
```
